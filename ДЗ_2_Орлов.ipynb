```python
#Завдання 2
num1=int(input('Enter first integer: ').strip())
num2=int(input('Enter second integer: ').strip())
if num1 > num2:
    print(f'{num1} більше, ніж {num2}')
elif num1 < num2:
    print(f'{num2} більше, ніж {num1}')
else:
    print(f'{num1} дорівнює {num2}')
```
