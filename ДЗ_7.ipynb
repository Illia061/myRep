```python
import pandas as pd

# Завантажте набір даних Пауерліфтинг
url = 'https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2019/2019-10-08/ipf_lifts.csv'
df_lift = pd.read_csv(url)
```

```python
#Знайдіть рекорди кожної статі та кожного підрозділу в кожній із 3 вправ

best3squat=df_lift.groupby(['sex', 'division']).agg({'name':'first','best3squat_kg': 'max'}).reset_index()
print(best3squat)

best3bench=df_lift.groupby(['sex', 'division']).agg({'name':'first','best3bench_kg': 'max'}).reset_index()
print(best3bench)

best3deadlift=df_lift.groupby(['sex', 'division']).agg({'name':'first','best3deadlift_kg': 'max'}).reset_index()
print(best3deadlift)
```

```python
#Порахуйте кількість перемог кожного учасника, беручи за перемогу 1-е місце
top_winner=df_lift[df_lift['place']=='1'].groupby(['name','sex', 'division']).size().reset_index(name='Count_win').sort_values('Count_win', ascending=False)
print(top_winner.head())
```

```python
#Використайте DataFrame з вправи 3, щоб показати для кожної комбінації підрозділу та статі, учасника з найбільшою кількістю перемог серед учасників
top_winner.groupby(['sex', 'division']).agg({'name':'first','Count_win': 'max'}).reset_index()
```
