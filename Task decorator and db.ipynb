```python
#Task 1.
def type_check(*types):
    '''
    Декоратор для перевірки відповідністі типів аргументів функції заданим типам

    :param *types: Задані типи аргументів.

    Повертає виняток, якщо типи не збігаються.
    '''
    def decorator(func):
        def wrapper(*args, **kwargs):
            for arg, arg_type in zip(args, types):
                if isinstance(arg, arg_type):
                    result = func(*args, **kwargs)
                    return result
                else:  raise TypeError(f"Очікувався тип {arg_type}, але отримано {type(arg)} для аргумента {arg}")
        return wrapper
    return decorator

# Приклад використання:

@type_check(int, int)
def add(a, b):
    return a + b

print(add(1.6, 2))  
```

```python
#Task 2.
import time
import requests

def delay(seconds):
    '''
    Декоратор для затримки функції на задану кількість секунд

    :param seconds: Задана кількість секунд.
    '''
    def decorator(func):
        def wrapper(*args, **kwargs):
            time.sleep(seconds)  # Затримуємо виконання на вказану кількість секунд
            '''
            Перевірка статус коду
            '''
            try:
                result = func(*args, **kwargs)  # Викликаємо оригінальну функцію

                if hasattr(result, 'status_code') and result.status_code != 200:
                    print(f"Увага: Отримано статус-код не 200: {result.status_code}")

                return result

            except requests.exceptions.RequestException as e:
                print(f"Помилка під час запиту: {e}")
                return None 

        return wrapper

    return decorator

# Приклад використання:

@delay(2)
def example_request():
    return requests.get("https/www.examle.com")

# Виклик функції з затримкою та перевіркою

result = example_request()
print(result)  # Друк результату або None у випадку помилки
```

```python
#SQL Task 1
'''
Скрипт створює базу данних з таблицею Person, with id: int, primary key, name: text, surname: text,
date_of_birth date
'''
# Connect with BD 
import sqlite3
from faker import Faker
# create a connection to the database
conn = sqlite3.connect('example1.db')

print(conn)
# create a table
conn.execute('''CREATE TABLE IF NOT EXISTS
            person (id INTEGER PRIMARY KEY, name TEXT, surname TEXT,
date_of_birth DATE)''')

# commit the changes
conn.commit()
conn.close()
```

```python
#SQL Task 2
'''
Скрипт заповнює таблицю Person на 15 осіб
'''
with sqlite3.connect('example1.db') as conn:
    fake = Faker()

    Faker.seed(42)


    for i in range(1, 16):
        conn.execute("INSERT INTO person (id, name, surname, date_of_birth ) VALUES (?, ?, ?, ?)", (i, fake.first_name(), fake.last_name(), fake.date_of_birth(minimum_age=18, maximum_age=65)))

    conn.commit()
```

```python
#SQL Task 3
'''
Скрипт виводить осіб молодших 1998-03-09
'''
with sqlite3.connect('example1.db') as conn:
    cursor = conn.cursor()
    cursor.execute("SELECT * FROM person WHERE date_of_birth>'1998-03-09' ORDER BY date_of_birth DESC")
    result = cursor.fetchall()
    
    for i in result:
        print(i)
```

```python
#SQL Task 4
'''
Скрипт виводить 5 найстарших осіб
'''
with sqlite3.connect('example1.db') as conn:
    cursor = conn.cursor()
    cursor.execute("SELECT * FROM person ORDER BY date_of_birth")
    result = cursor.fetchmany(5)
    
    for i in result:
        print(i)
```

```python
#SQL Task 5
'''
Скрипт виводить осіб в яких довжина імені більша за довжину прізвища
'''
with sqlite3.connect('example1.db') as conn:
    cursor = conn.cursor()
    cursor.execute("SELECT * FROM person WHERE length(name)>length(surname) ORDER BY id")
    result = cursor.fetchall()
    
    for i in result:
        print(i)
```
