```python
import warnings

warnings.simplefilter(action='ignore', category=FutureWarning)

import seaborn as sns

dm = sns.load_dataset("diamonds")

#Середня ціна діамантів
avg_df=round(dm["price"].mean(), 2)

#Середня к-ть каратів по кожному огрануванню
avg_carat=dm.groupby(["cut"]).mean(["carat"])

#найпоширеніший колір
moda_color=dm["color"].mode()

#розподіл діамантів по кольорам
dm["color"].hist()

#відсоток діамантів з огрануванням Fair
percent_fair=round((dm["cut"][dm["cut"]=="Fair"].count()/dm["cut"].count())*100, 2)

#відсоток діамантів з огрануванням Ideal
percent_ideal=round((dm["cut"][dm["cut"]=="Ideal"].count()/dm["cut"].count())*100, 2)

#кількість діамантів Premium у кожному кольорі
dm["color"][dm["cut"]=="Premium"].hist()
```
